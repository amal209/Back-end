```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import pandas as pd
from selenium.webdriver.common.by import By
```

# Chrome webDriver

```python
#Chrome webDriver
driver = webdriver.Chrome('C:/Users/Amal/Desktop/chromedriver')
```

# Scrapping Population by Continents

## 1. Africa

```python
#Access Website Via Python
driver.get('https://www.worldometers.info/world-population/africa-population/')
```

```python
#alll
all = driver.find_elements(by=By.XPATH, value='//table[@class="table table-striped table-bordered table-hover table-condensed table-list"]/tbody/tr/td')
all_list = []

for y in range(len(all)):
    #print(years[y].text)
    all.append(all[y].text)
#print(years_list)

#years
years = driver.find_elements(by=By.XPATH, value='//table[@class="table table-striped table-bordered table-hover table-condensed table-list"]/tbody/tr/td[1]')
years_list = []
for y in range(len(years)):
    years_list.append(years[y].text)
print(years_list)

#population
population = driver.find_elements(by=By.XPATH, value='//table[@class="table table-striped table-bordered table-hover table-condensed table-list"]/tbody/tr/td[2]')
population_list = []
for p in range(len(population)):
    population_list.append(population[p].text)
print(population_list)

#median age
medianAge = driver.find_elements(by=By.XPATH, value='//table[@class="table table-striped table-bordered table-hover table-condensed table-list"]/tbody/tr/td[6]')

medianAge_list = []
for ma in range(len(medianAge)):
    medianAge_list.append(medianAge[ma].text)
print(medianAge_list)

#Urban Population
urbanPopulation = driver.find_elements(by=By.XPATH, value='//table[@class="table table-striped table-bordered table-hover table-condensed table-list"]/tbody/tr/td[10]')

urbanPopulation_list = []
for up in range(len(urbanPopulation)):
    urbanPopulation_list.append(urbanPopulation[up].text)
print(urbanPopulation_list)
```

### dataframe 

```python
africa_df = pd.DataFrame(list(zip(years_list, population_list, medianAge_list, urbanPopulation_list)),
               columns =['Year', 'Population','Median Age','Urban Population'])
africa_df
```

### Splitting dataframe into Population of Africa (2020 and historical) AND Africa Population Forecast

```python
historical_african_population_df = africa_df.iloc[:18,:]
africa_population_forecast_df = africa_df.iloc[18:,:]
```
